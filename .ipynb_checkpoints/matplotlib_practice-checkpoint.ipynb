```python
# Displayed plots inline within JN, directly below the code cell that produced it
%matplotlib inline

# "from matplotlib depencency, import pyplot module (to create chart w/ features), then rename it plt"
import matplotlib.pyplot as plt

# add the data:

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# ------MATLAB Method-------------

# Create the plot.
# declare the parameters for marker, line color, width
# add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#---Object-Oriented Method---------------

# Create the figure with the current axes
fig, ax = plt.subplots()

# define the axes
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a label legend for the city of Boston.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add a y-axis limit.
ax.set_ylim(0, 45)
# Add grid lines.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#___MATLAB Method______BAR CHARTS_______________________________

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#________Object-Oriented Approach______BAR CHARTS____________

# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis, color="cyan", label='Chicago')
# Add the legend.
ax.legend()
# Create labels for the x and y axes.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.invert_yaxis()
ax.set_title("PyBar Fare by Month")
```

```python
# Bubble Chart Using the Object-Oriented Method

# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, c="xkcd:sky blue", s=[i * 3 for i in x_axis], alpha=.8, edgecolors="black", linewidth=2, label='Boston')
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.

# Add a legend for the city of Boston.
ax.legend()

# Add a title.
ax.set_title("PyBer Fare by Month")

# Switch the axis so that the Fare($) data is on the x-axis.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

# Add a limit for the x-axis data.
ax.set_xlim(0, 45)

# Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```
